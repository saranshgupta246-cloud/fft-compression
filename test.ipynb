```python
import sys; sys.path.append('src')
from fft_core import compress_image_rgb
from utils import load_image, to_pil
from metrics import full_report
import matplotlib.pyplot as plt

img = load_image("assets/sample.jpg")
result = compress_image_rgb(img, keep_fraction=0.1)
report = full_report(img, result, 0.1)
print(report)

fig, ax = plt.subplots(1, 2, figsize=(12,5))
ax[0].imshow(img.astype('uint8')); ax[0].set_title("Original")
ax[1].imshow(result); ax[1].set_title("Compressed")
plt.show()
```
